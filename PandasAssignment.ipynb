```python
#Create a pandas DataFrame from the CSV and from Python lists/dicts.
import pandas as pd
df1=pd.read_csv("sample.csv")  
one = pd.DataFrame({ 'transaction_id': [1006] , 'store_id': [2], 'store_name':
["Suburb"], 'date':['2025-10-23'], 'product_id':[502], 'product_name':['Pen'],'quantity':[1], 'unit_price':[0.80], 'total':[0.80], 'customer_id':[2003], 'promo_code':["DISC10"] } )
df=pd.concat([df1,one])
print(df.tail())
```

```python
#Filter and subset rows/columns (e.g., sales for a store, date ranges, high-value transactions).
store3df=df[(df['store_id']==3)&(df['total']>5.0)&(df['date']>='2025-10-03')&(df['date']<='2025-10-26')]
print(store3df)
print("total rows that satisfy the condition ",len(store3df))
```

```python
df['customer_id']=df['customer_id'].astype(int)
df.info()
```

```python
#Compute descriptive statistics and group summaries (totals by store, average basket).
print(df.describe())
totalsByStore=df.groupby('store_id')['total'].sum()
print(totalsByStore)
averageBasket=df.groupby('store_id')['total'].mean()
print(averageBasket)
store_summary = df.groupby('store_id').agg(
    total_sales = ('total', 'sum'),
    avg_basket = ('total', 'mean'),
    transaction_count = ('total', 'count')
)

print(store_summary)
```

```python
#Perform simple cleaning (handle missing values, fix data types, drop duplicates).
#count missing values
df.isnull().sum()
```

```python
#filling missing values of customer_id with mean
meanValue=int(df['customer_id'].mean())
df['customer_id']=df['customer_id'].fillna(meanValue)
#filling missing values of promo_code with None
df['promo_code']=df['promo_code'].fillna("None")
#chceck for missing values count
print(df.isnull().sum())
```

```python
#check for duplicates
print(df.duplicated().sum())
```

```python
dfClean=df
dfClean
```

```python
#Build a mini ETL pipeline: read CSV → clean & transform → output JSON.(Reading CSV, cleaning&transforming is done in the above steps)
dfClean.to_json("output.json", orient="records", indent=4)
```
